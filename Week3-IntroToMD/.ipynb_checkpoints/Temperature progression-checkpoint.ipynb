```python
import matplotlib.pyplot as plt
import numpy as np

# Load the data
data = np.loadtxt('Analysis/temperature.xvg', comments=['@', '#'])
time = data[:, 0]  # Assuming the first column is time
temperature = data[:, 1]  # Assuming the second column is temperature

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(time, temperature, label='Temperature (K)', color='blue')
plt.title('Temperature Progression Over Time')
plt.xlabel('Time (ps)')
plt.ylabel('Temperature (K)')
plt.xlim(left=0)  # Start x-axis from 0
plt.grid()
plt.legend()
plt.savefig('Analysis/temperature_progression.png')  # Save the plot as an image
plt.show()
```
