```python
import matplotlib.pyplot as plt
import numpy as np

# Load the data
data = np.loadtxt('Analysis/density.xvg', comments=['@', '#'])
time = data[:, 0]  # Assuming the first column is time
density = data[:, 1]  # Assuming the second column is density

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(time, density, label='Density (g/cm³)', color='green')
plt.title('Density Progression Over Time')
plt.xlabel('Time (ps)')
plt.ylabel('Density (g/cm³)')
plt.xlim(left=0)  # Start x-axis from 0
plt.grid()
plt.legend()
plt.savefig('Analysis/density_progression.png')  # Save the plot as an image
plt.show()
```
