```python
import matplotlib.pyplot as plt
import numpy as np

# Load the data
data = np.loadtxt('Analysis/pressure.xvg', comments=['@', '#'])
time = data[:, 0]  # Assuming the first column is time
pressure = data[:, 1]  # Assuming the second column is pressure

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(time, pressure, label='Pressure (bar)', color='red')
plt.title('Pressure Progression Over Time')
plt.xlabel('Time (ps)')
plt.ylabel('Pressure (bar)')
plt.xlim(left=0)  # Start x-axis from 0
plt.grid()
plt.legend()
plt.savefig('Analysis/pressure_progression.png')  # Save the plot as an image
plt.show()
```
