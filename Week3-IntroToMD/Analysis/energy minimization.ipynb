```python
import matplotlib.pyplot as plt
import numpy as np

# Load the potential energy data
data = np.loadtxt('Analysis/potential.xvg', comments=['@', '#'])

# Plot the potential energy
plt.plot(data[:, 0], data[:, 1])
plt.xlabel('Step')
plt.ylabel('Potential Energy (kJ/mol)')
plt.title('Energy Minimization')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Load the potential energy data
data = np.loadtxt('Analysis/potential.xvg', comments=['@', '#'])

# Plot the potential energy
plt.plot(data[:, 0], data[:, 1])
plt.xlabel('Step')
plt.ylabel('Potential Energy (kJ/mol)')
plt.title('Energy Minimization')

# Set x-axis limit to start at 0
plt.xlim(left=0)

# Display the plot
plt.show()
```
