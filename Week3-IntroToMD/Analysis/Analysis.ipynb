```python
import numpy as np
import matplotlib.pyplot as plt

# Function to read .xvg files
def read_xvg(filename):
    data = np.loadtxt(filename, comments=['@', '#'])  # Load the data while ignoring comments
    return data[:, 0], data[:, 1]  # Return time and RMSD values

# Read RMSD data
time1, rmsd1 = read_xvg('rmsd.xvg')      # Adjust path if necessary
time2, rmsd2 = read_xvg('rmsd_xtal.xvg')  # Adjust path if necessary

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(time1, rmsd1, label='RMSD (Minimized Structure)', color='blue')
plt.plot(time2, rmsd2, label='RMSD (Crystal Structure)', color='orange')

# Adding titles and labels
plt.title('RMSD Analysis')
plt.xlabel('Time (ns)')
plt.ylabel('RMSD (nm)')
plt.legend()
plt.grid()

# Show the plot
plt.show()
```

```python
# Save the figure
plt.savefig('rmsd_plot.png')  # Save the plot as a PNG file
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Function to read .xvg files
def read_xvg(filename):
    data = np.loadtxt(filename, comments=['@', '#'])  # Load the data while ignoring comments
    return data[:, 0], data[:, 1]  # Return time and Rg values

# Read the radius of gyration data
time, rg = read_xvg('gyrate.xvg')  # Adjust path if necessary

# Plotting the radius of gyration
plt.figure(figsize=(10, 6))
plt.plot(time, rg, label='Radius of Gyration (Rg)', color='green')

# Adding titles and labels
plt.title('Radius of Gyration Over Time')
plt.xlabel('Time (ns)')
plt.ylabel('Radius of Gyration (nm)')
plt.legend()
plt.grid()

# Show the plot
plt.show()
```
