```python
import mdtraj as md
import numpy as np
import matplotlib.pyplot as plt

# Load the unwrapped trajectory
traj = md.load('unwrapped.xtc', top='step5_1.gro')  # Use the .gro file as topology

# Print the topology information to check atom names
print("Topology Information:")
print(traj.topology)

# Select atoms for Water Oxygen, Sodium, and Chloride
# Modify the selection strings based on the printed topology
water_oxygens = traj.topology.select('name OW')  # Adjust if needed
sodium = traj.topology.select('name SOD')         # Adjust if needed
chloride = traj.topology.select('name CLA')       # Adjust if needed

# Ensure selections are valid and not empty
if len(water_oxygens) == 0:
    raise ValueError("No water oxygen indices found.")
if len(sodium) == 0:
    raise ValueError("No sodium indices found.")
if len(chloride) == 0:
    raise ValueError("No chloride indices found.")

# Function to compute Mean Squared Displacement (MSD)
def compute_msd(traj, atom_indices):
    # Slice the trajectory to get only the atoms we're interested in
    sliced_traj = traj.atom_slice(atom_indices)

    # Get the positions of the atoms at all frames
    positions = sliced_traj.xyz

    # Calculate the center of mass for each frame
    com = np.mean(positions, axis=1)  # Average over the atom indices

    # Calculate the squared displacements
    msd = np.mean(np.square(com - com[0]), axis=0)

    return msd

# Compute MSD for each group
msd_water = compute_msd(traj, water_oxygens)   # MSD for water oxygens
msd_sodium = compute_msd(traj, sodium)         # MSD for sodium ions
msd_chloride = compute_msd(traj, chloride)     # MSD for chloride ions

# Plotting the MSD results
plt.figure(figsize=(10, 6))
plt.plot(msd_water, label='Water Oxygens')
plt.plot(msd_sodium, label='Sodium Ions')
plt.plot(msd_chloride, label='Chloride Ions')
plt.xlabel('Time (ps)')
plt.ylabel('Mean Squared Displacement (nm^2)')
plt.title('MSD vs Time')
plt.legend()
plt.grid()
plt.savefig('msd_plot.png')
plt.show()
```
