```python
import os
import numpy as np
import matplotlib.pyplot as plt

# Create the 'Figures' directory if it doesn't exist
if not os.path.exists('Figures'):
    os.makedirs('Figures')

# Example arrays (replace with your actual data)
time = np.linspace(0, 50, num=500)  # Example time points (0 to 50 ns)
end_to_end_distance = np.random.rand(500) * 3  # Example values for end-to-end distance (in nm)
radius_of_gyration = np.random.rand(500) * 2  # Example values for radius of gyration (in nm)

# Create plots
plt.figure(figsize=(12, 6))

# Plot End-to-End Distance
plt.subplot(1, 2, 1)
plt.plot(time, end_to_end_distance, label='End-to-End Distance', color='blue')
plt.xlabel('Time (ns)')
plt.ylabel('Distance (nm)')
plt.title('End-to-End Distance vs. Time')
plt.legend()
plt.grid()

# Plot Radius of Gyration
plt.subplot(1, 2, 2)
plt.plot(time, radius_of_gyration, label='Radius of Gyration', color='orange')
plt.xlabel('Time (ns)')
plt.ylabel('Radius of Gyration (nm)')
plt.title('Radius of Gyration vs. Time')
plt.legend()
plt.grid()

# Adjust layout and save the figure
plt.tight_layout()
plt.savefig('Figures/end_to_end_radius_of_gyration.png')  # Save as PNG
plt.show()  # Display the plots
```
