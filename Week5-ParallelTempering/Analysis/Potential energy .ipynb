```python
import matplotlib.pyplot as plt
import numpy as np

# Load potential energy data for each replica
energy_T300 = np.loadtxt('T300/potential_energy.xvg', comments=['#', '@'])
energy_T400 = np.loadtxt('T400/potential_energy.xvg', comments=['#', '@'])
energy_T600 = np.loadtxt('T600/potential_energy.xvg', comments=['#', '@'])

# Plot histograms
plt.hist(energy_T300[:, 1], bins=50, alpha=0.5, label='T300', density=True)
plt.hist(energy_T400[:, 1], bins=50, alpha=0.5, label='T400', density=True)
plt.hist(energy_T600[:, 1], bins=50, alpha=0.5, label='T600', density=True)

plt.xlabel('Potential Energy (kJ/mol)')
plt.ylabel('Density')
plt.legend()
plt.title('Potential Energy Histogram for Each Temperature')
plt.show()
```
