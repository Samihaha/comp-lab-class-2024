```python
import numpy as np
import matplotlib.pyplot as plt

# Load dihedral data (phi and psi)
dihedrals = np.loadtxt('dihedrals.xvg', comments=['#', '@'])

# Compute 2D histogram (free energy surface)
phi = dihedrals[:, 0]  # Assuming phi is in column 1
psi = dihedrals[:, 1]  # Assuming psi is in column 2
H, xedges, yedges = np.histogram2d(phi, psi, bins=50, range=[[-180, 180], [-180, 180]])

# Free energy in kT (arbitrary units, shift to minimum 0)
free_energy = -np.log(H + 1e-10)
free_energy -= free_energy.min()

# Plot the free energy surface
plt.contourf(xedges[:-1], yedges[:-1], free_energy.T, levels=np.linspace(0, 6, 50))
plt.colorbar(label='Free Energy (kT)')
plt.xlabel('Phi (degrees)')
plt.ylabel('Psi (degrees)')
plt.title('Free Energy Surface (T=300K)')
plt.show()
```
