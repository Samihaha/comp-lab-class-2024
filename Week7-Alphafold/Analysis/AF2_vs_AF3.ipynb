```python
import mdtraj as md
import numpy as np
import matplotlib.pyplot as plt

# Load the models
af2_model = md.load('/home/sa8200/comp-lab-class/comp-lab-class-2024/Week7-Alphafold/Inputs/AF3_vs_AF2/AF2_Actin_Profilin/Actin_Profilin_923b1_0/Actin_Profilin_923b1_0_unrelaxed_rank_005_alphafold2_multimer_v3_model_2_seed_000.pdb')  # Replace with the path to your AF2 model
af3_model = md.load('/home/sa8200/comp-lab-class/comp-lab-class-2024/Week7-Alphafold/Inputs/AF3_vs_AF2/2btf.pdb')  # Replace with the path to your AF3 model

# Select the backbone atoms of Actin (assuming chain A, or chainid 0)
af2_actin_indices = af2_model.top.select('chainid 0 and backbone')
af3_actin_indices = af3_model.top.select('chainid 0 and backbone')

# Ensure both selections have the same number of atoms
if len(af2_actin_indices) > len(af3_actin_indices):
    af2_actin_indices = af2_actin_indices[:len(af3_actin_indices)]
elif len(af3_actin_indices) > len(af2_actin_indices):
    af3_actin_indices = af3_actin_indices[:len(af2_actin_indices)]

# Align the models based on the backbone atoms
af2_model.superpose(af3_model, atom_indices=af2_actin_indices)

# Calculate distances between corresponding backbone atoms
distances = np.sqrt(np.sum((af2_model.xyz[0, af2_actin_indices] - af3_model.xyz[0, af3_actin_indices])**2, axis=1))

# Plot the distances
plt.figure(figsize=(10, 5))
plt.bar(range(len(distances)), distances, color='skyblue')
plt.xlabel("Residue Number (Backbone Atoms)")
plt.ylabel("Distance (nm)")
plt.title("Backbone Atom Distance between AF2 and AF3 Predictions")
plt.show()
```
