```python
import mdtraj as md
import glob

# Update with correct path
pdb_files = glob.glob('/home/sa8200/comp-lab-class/comp-lab-class-2024/Week7-Alphafold/Inputs/mini_protein_arp-est/*.pdb')
traj = md.load(pdb_files)
```

```python
# Align all frames to the first frame
traj.superpose(traj[0])

# Compute per-atom RMSF
rmsf_per_atom = md.rmsf(traj, traj[0])

# Calculate average RMSF per residue
residue_rmsf = []
for res in traj.topology.residues:
    if res.is_protein:
        # Get indices of atoms in this residue
        atom_indices = [atom.index for atom in res.atoms]
        # Average the RMSF of all atoms in the residue
        residue_rmsf.append(rmsf_per_atom[atom_indices].mean())

# Plot RMSF per residue
plt.figure(figsize=(10, 5))
plt.bar(range(len(residue_rmsf)), residue_rmsf, color='lightcoral')
plt.xlabel("Residue Index")
plt.ylabel("RMSF (nm)")
plt.title("RMSF per Residue for mini protein arp-est")
plt.show()
```
